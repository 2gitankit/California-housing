```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
dt = fetch_california_housing()
```

```python
print(dt)
```

```python
ch_df = pd.DataFrame(dt.data,columns=dt.feature_names)
```

```python
ch_df
```

```python
ch_df.shape
```

```python
ch_df['target']=dt['target']
```

```python
ch_df
```

```python
sns.heatmap(ch_df.corr(),annot=True,cmap="coolwarm")
```

```python
#Observations 
#MedInc is directly proportional to target.
#AveRooms and AveBedrms are colinear in nature. 
```

# Multi - Linear Regression

```python
#Splitting the data 
x = ch_df.drop('target', axis = 1)
y = ch_df['target']
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
r2_s = r2_score(y_test, y_pred)
mse_err = mean_squared_error(y_test, y_pred)

print(r2_s, mse_err)
```

# Result 
R2 Square - 0.59 &
Mean Squared Error - 0.53
